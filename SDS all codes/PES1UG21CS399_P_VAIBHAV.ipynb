```python
df.columns=['Timestamp','Gender','Your study year','Which department','How many electronic devices do you use?','What is your recent latest CGPA ?','Is there an increase in your CGPA?',	'Average hour you spent using electronic devices for academic purposes in a day (hours).','	What is the purpose of buying the device?']
df.info()
```

```python
df.isnull().values.any()
df.isnull().sum()
```

```python
df['Average hour you spent using electronic devices for academic purposes in a day (hours).'] = df['Average hour you spent using electronic devices for academic purposes in a day (hours).'].fillna(df['Average hour you spent using electronic devices for academic purposes in a day (hours).'].median())
df['How many electronic devices do you use?'] = df['How many electronic devices do you use?'].fillna(df['How many electronic devices do you use?'].median())

df
```
