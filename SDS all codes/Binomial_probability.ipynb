```python

import math
import matplotlib.pyplot as plt
import numpy as np
from math import factorial as fact

def combination(n,x):
    comb=fact(n)/(fact(x)*fact(n-x))
    return comb
def binom(n,p):
    pmf=[]
    x=np.arange(n+1)
    for k in range(0, n+1):
        prob=combination(n,k)*math.pow(p,k)*math.pow(1-p,n-k)
        pmf.append(prob)
    
    plt.bar(x,pmf)
    plt.xlabel("Number of successes")
    plt.ylabel("P(x)")
    plt.title("Probability Histogram")
    plt.show()
    
binom(50,0.3)
```

```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

p=[]
n=20
x=np.arange(n+1)
for i in range(0,n+1):
    ax=scipy.stats.binom.pmf(i,n,0.8)
    p.append(ax)
    
plt.bar(x,p,width=1)
plt.xlabel("No. of successes")
plt.ylabel("P(x)")
plt.xticks(x+0.5,x)
plt.title("Probabilty Histogram")
plt.show()
```
