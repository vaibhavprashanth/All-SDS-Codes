```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import sample
from statistics import mean

def plot(arr, N, n_samples):
    x = []
    for i in range(1, n_samples):
        #to find N samples from the arr
        smp = sample(arr, N)
        mu = mean(smp)
        x.append(mu)
    plt.hist(x)
    plt.show()


df = pd.read_csv("top_100_anime_movie.csv")
arr = df['rating']
arr = np.array(arr)
arr = sorted(arr)
print(arr)

#variations
plot(arr, 1, 10)
plot(arr, 2, 20)
plot(arr, 5, 100)
#so as number of samples(n_sample) increases the distribution becomes normal
#so as sample size increases the flatness of the distribution decreses
```
