```python
import math
import matplotlib.pyplot as plt
import numpy as np

def poisson(rate):
    x=0
    pmf=[]
    prob=1
    while prob > 0.0000000000000000000000000001:
        prob=math.exp(-rate)*pow(rate,x)/math.factorial(x)
        pmf.append(prob)
        x=x+1
    n=np.arange(0,x)
    plt.bar(n,pmf,width=1)
    plt.show()

poisson(2)
poisson(30)
```

```python
#Area in an interval between two values a and b
from scipy.stats import norm
def prob(a,b):
    z1=(a-mu)/sigma
    z2=(b-mu)/sigma
    print("Finding P(",z1,"< z <",z2,")")
    print("Area to left of a: ",norm.cdf(z1))
    print("Area to left of b: ",norm.cdf(z2))
    print("Area between a and b: ",norm.cdf(z2)-norm.cdf(z1))

mu=50
sigma=5
prob(44.5,55.5)
```
